1. Se consideră cheia publică RSA cu modulul pe 128 biți:

N=234841136411758273000763594354834942653

e=65537

Factorizați modulul, i.e. determinați valorile p și q. Calculați apoi coeficientul de decriptare d.


```python
from sympy import factorint, mod_inverse

N = 234841136411758273000763594354834942653
e = 65537

# Factorizam modulul
p, q = factorint(N)

# Calculam d
phi = (p-1) * (q-1)
d = mod_inverse(e, phi)

print("p =", p)
print("q =", q)
print("Coef d =", d)
```
